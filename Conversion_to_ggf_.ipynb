```python
from unsloth import FastLanguageModel
```

```python
model, tokenizer = FastLanguageModel.from_pretrained("vinci00/scrapegraphcompanion_finetuning_demo-adapter", max_seq_length=4096,
        dtype=None,
        load_in_4bit=True)
```

```python
model.push_to_hub_gguf(
    f"vinci00/scrapegraphcompanion_finetuning_GGUF",
    tokenizer,
    quantization_method="q4_k_m",
    token="HF_TOKEN"
)
```

```python
model.push_to_hub_gguf(
    f"vinci00/scrapegraphcompanion_finetuning_GGUF",
    tokenizer,
    quantization_method="q5_k_m",
    token="HF_TOKEN"
)
```
